```python
iris = load_iris()
```

```python
dir(iris)
```

```python
iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df
```

```python
iris_df['target'] = iris.target
iris_df
```

```python
iris_df['target_species'] = iris_df.target.apply(lambda x : iris.target_names[x])
iris_df
```

```python
groups = iris_df.groupby('target_species')
```

```python
groups.get_group("virginica")
```

```python
groups.get_group("setosa")
```

```python
iris_df['target_species'].unique()
```

```python
df0 = iris_df[iris_df.target_species == 'virginica']
df1 = iris_df[iris_df.target_species == 'setosa']
df2 = iris_df[iris_df.target_species == 'versicolor']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'],color="green",marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color="blue",marker='.')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'],color="green",marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'],color="blue",marker='.')
```

```python
x = iris_df.drop(['target_species','target'],axis='columns')
y = iris_df['target_species']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
#gamma
model_g = SVC(gamma=10)
model_g.fit(x_train, y_train)
model_g.score(x_test, y_test)
```

```python
#kernel
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(x_train, y_train)
model_linear_kernal.score(x_test, y_test)
```
